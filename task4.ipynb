# Recharge Validation

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount >= 50 and amount in valid_plans:
        return True
    return False

while True:
    amt = int(input("Enter recharge amount: ₹"))

    if validate_recharge(amt):
        print("Recharge Successful ")
        break
    else:
        print("Invalid Recharge! Try again ")
```

# Inventory Reorder

```python
def check_inventory(stock):
    for product, qty in stock.items():
        if qty < 15:
            print(product, ": Reorder Alert ")
        else:
            print(product, ": Stock OK")

inventory = {
    "Keyboard": 10,
    "Mouse": 25,
    "Monitor": 8,
    "Printer": 20
}

check_inventory(inventory)
```

# Result Processing

```python
def student_result(marks):
    total = 0
    for m in marks:
        total += m

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

marks = [60, 55, 40, 65]
print("Result:", student_result(marks))
```

# Fare Estimator with Retry

```python
def calculate_fare(km, peak):
    fare = 50 + (km * 12)
    if peak == "yes":
        fare += fare * 0.25
    return fare

while True:
    km = float(input("Enter distance (km): "))
    peak = input("Peak hour? (yes/no): ").lower()

    print("Total Fare: ₹", calculate_fare(km, peak))

    retry = input("Calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

# Employee Attendance Eligibility

```python
def attendance_check(attendance):
    present = 0
    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible "
    else:
        return "Not Eligible"

attendance = ["P", "A", "P", "P", "P", "A"]
print(attendance_check(attendance))
```

# Password Strength

```python
def check_password(password):
    has_digit = False
    has_special = False

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password "
    else:
        return "Weak Password "

pwd = input("Enter password: ")
print(check_password(pwd))
```
