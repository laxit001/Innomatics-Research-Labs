# Unique word 

```python
sentence = "python is easy and python is powerful"

words = sentence.split()
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

# Higest Salary

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = max(employees, key=employees.get)
print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

# Max and Min 

```python
numbers = [45, 22, 89, 10, 66]
maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

# Count Product

```python
prices = [450, 1200, 899, 1500, 300]

count = 0
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

# Attendence percentage

```python
attendance = ["P", "P", "A", "P", "P"]

present_count = attendance.count("P")
percentage = (present_count / len(attendance)) * 100

print("Attendance Percentage:", percentage)
```

# Duplicate phone no.

```python
phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

# Count Charecter

```python
text = "python"

char_freq = {}

for ch in text:
    char_freq[ch] = char_freq.get(ch, 0) + 1

print(char_freq)
```

# Key Exists in a Dictionary

```python
employees = {"Ravi": 75000, "Anita": 68000}

if "Ravi" in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

# Average Marks

```python
marks = [70, 75, 80, 65]

average = sum(marks) / len(marks)

print("Average Marks:", average)
```
