# Smart Parking Lot

```python
def parking_lot_manager(capacity, logs):
    parked = 0
    peak = 0

    for action in logs:
        if action == "IN":
            parked += 1
        elif action == "OUT" and parked > 0:
            parked -= 1

        peak = max(peak, parked)

    print("Currently parked vehicles:", parked)

    if parked > capacity:
        print("Parking status: full")
    else:
        print("Parking status: available")

capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
parking_lot_manager(capacity, logs)
```

# Online Food Delivery

```python
def delivery_time_estimator(distance, traffic, weather):
    time = distance * 5   # base time: 5 minutes per km

    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 8

    if weather == "Rainy":
        time += 10
    elif weather == "Stormy":
        time += 20

    print("Estimated delivery time:", time, "minutes")
delivery_time_estimator(8, "High", "Rainy")
```

# Movie Theatre Seat

```python
def seat_occupancy(total_seats, booked_seats):
    booked = len(booked_seats)
    occupancy = (booked / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: housefull")
    elif occupancy >= 75:
        print("Show Status: almost full")
    else:
        print("Show Status: seats available")

total_seats = 200
booked_seats = [1] * 150
seat_occupancy(total_seats, booked_seats)
```

# Cloud Server Load

```python
def server_load_classifier(cpu_readings):
    average = sum(cpu_readings) // len(cpu_readings)

    print("Average CPU Load:", average, "%")

    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average < 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

cpu = [45, 60, 70, 85, 90]
server_load_classifier(cpu)
```

# Smart Classroom Resource

```python
def classroom_resource_monitor(resources):
    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused resources:", ", ".join(overused))
        print("Energy alert: Yes")
    else:
        print("Overused resources: None")
        print("Energy alert: No")
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
classroom_resource_monitor(usage)
```

# Online Event Registration

```python
def event_registration_manager(capacity, registrations):
    if registrations <= capacity:
        print("Confirmed registrations:", registrations)
        print("Waitlisted users: 0")
        print("Registration status: Open")
    else:
        print("Confirmed registrations:", capacity)
        print("Waitlisted users:", registrations - capacity)
        print("Registration status: Closed")
event_registration_manager(100, 105)
```
