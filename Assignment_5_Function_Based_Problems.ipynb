# Assignment 5 – Real-World Function-Based Problem Solving

```

# Problem 1: Social Media – Post Engagement Analyzer
def analyze_engagement(likes):
    total_likes = sum(likes)
    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")

analyze_engagement([200, 300, 250, 300])
```

```

# Problem 2: Healthcare – Medicine Stock Alert System
def medicine_stock(stock):
    status = "Low Stock Alert" if stock < 10 else "Stock Sufficient"
    print(f"Medicine Stock: {stock}")
    print(f"Status: {status}")

medicine_stock(6)
```

```

# Problem 3: Agriculture – Rainfall Adequacy Checker
def rainfall_checker(rainfall, required):
    avg = sum(rainfall) / len(rainfall)
    status = "Adequate Rainfall" if avg >= required else "Inadequate Rainfall"
    print(f"Average Rainfall: {avg}")
    print(f"Rainfall Status: {status}")

rainfall_checker([70, 75, 72, 71], 70)
```

```

# Problem 4: Social Media – Duplicate Account Detection
def detect_duplicates(usernames):
    has_duplicates = len(usernames) != len(set(usernames))
    print("Duplicate Accounts Found:", "Yes" if has_duplicates else "No")

detect_duplicates(["user1", "user2", "user1"])
```

```

# Problem 5: Healthcare – Appointment Eligibility Checker
def check_eligibility(age):
    status = "Eligible" if age >= 18 else "Not Eligible"
    print(f"Patient Age: {age}")
    print(f"Eligibility Status: {status}")

check_eligibility(21)
```

```

# Problem 6: Agriculture – Premium Crop Price Filter
def premium_crops(prices):
    premium = [p for p in prices if p > 2000]
    print(f"Premium Crops: {premium}")

premium_crops([1500, 2500, 1800, 3200])
```

```

# Problem 7: System Monitoring – Application Health Checker
def system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    print(f"Error Count: {errors}")
    print(f"System Status: {status}")

system_health(7)
```

```

# Problem 8: Banking – Daily Transaction Limit Checker
def transaction_check(amount):
    limit = 50000
    status = "Approved" if amount <= limit else "Rejected"
    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")

transaction_check(60000)
```

```

# Problem 9: E-Learning – Student Attendance Eligibility System
def attendance_checker(attendance):
    percentage = (sum(attendance) / len(attendance)) * 100
    status = "Eligible" if percentage >= 75 else "Not Eligible"
    print(f"Attendance Percentage: {percentage}")
    print(f"Exam Eligibility: {status}")

attendance_checker([1,1,1,0,1])
```

```

# Problem 10: Smart Electricity Bill Analyzer
def electricity_bill(units):
    bill = 0
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage

bill, usage = electricity_bill(250)
print("Total Bill:", bill)
print("Usage Status:", usage)
```
