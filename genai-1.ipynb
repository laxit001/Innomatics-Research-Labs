## Q1. What is the difference between "Love" and "love" in NLP?

In Natural Language Processing (NLP), **"Love"** and **"love"** may be treated as two different tokens if case normalization is not applied.

For example:
- `"Love"` may appear at the beginning of a sentence.
- `"love"` may appear in the middle of a sentence.

If the text is not converted to lowercase, the NLP model may store both words separately in its vocabulary. This increases redundancy and makes the vocabulary unnecessarily larger.

### Example:
- `"Love is beautiful"`
- `"I love coding"`

Without lowercasing:
- Tokens = `["Love", "love"]`

With lowercasing:
- Tokens = `["love", "love"]`

### Conclusion:
To maintain consistency and reduce duplicate word forms, text is usually converted to lowercase during preprocessing.

## Q2. What happens if stopwords are not removed?

Stopwords are very common words such as:
- is
- the
- am
- and
- of
- to
- in

If stopwords are not removed, the dataset may contain too many words that do not contribute much meaning in certain NLP tasks.

### Effects of not removing stopwords:
- Increases text size
- Increases vocabulary size
- Adds unnecessary noise
- May reduce efficiency of machine learning models
- Important words may get less importance

### Example:
Sentence:
- `"This is the best course"`

Without stopword removal:
- `["this", "is", "the", "best", "course"]`

With stopword removal:
- `["best", "course"]`

### Conclusion:
If stopwords are not removed, the model may spend time processing words that are not always useful. However, whether stopwords should be removed depends on the NLP task.

## Q3. Provide two real-world scenarios where removing stopwords can be harmful.

Although stopword removal is useful in many NLP tasks, in some cases it can damage the meaning of the sentence.

---

### 1. Sentiment Analysis

Sentence:
- `"I am not happy"`

If stopwords like `"not"` are removed:
- `"I happy"`

This completely changes the meaning of the sentence.

Original meaning:
- Negative sentiment

After removing `"not"`:
- Positive or unclear sentiment

So in sentiment analysis, removing stopwords carelessly can lead to wrong predictions.

---

### 2. Question Answering / Chatbots

Sentence:
- `"What is the time now?"`

If stopwords are removed aggressively:
- `"time now"`

This removes the sentence structure and some contextual meaning.

In question-answering systems and chatbots, such words may help the system understand:
- intent
- sentence structure
- grammatical meaning

---

### Conclusion:
Stopword removal is not always safe. In tasks where context, negation, or grammar matters, stopwords should be handled carefully.

## Q4. Explain the difference between stemming and lemmatization.

Stemming and lemmatization are both text normalization techniques used in NLP to reduce words to their base or root form.

---

### 1. Stemming

Stemming removes word endings using simple rules, often without considering grammar or dictionary meaning.

#### Examples:
- `"playing"` → `"play"`
- `"studies"` → `"studi"`
- `"running"` → `"run"`

Stemming is:
- fast
- simple
- less accurate

Sometimes it produces words that are not valid dictionary words.

---

### 2. Lemmatization

Lemmatization converts a word into its correct dictionary base form using linguistic and grammatical rules.

#### Examples:
- `"playing"` → `"play"`
- `"studies"` → `"study"`
- `"better"` → `"good"` (depending on context)

Lemmatization is:
- slower
- more accurate
- linguistically correct

---

### Key Difference

| Feature | Stemming | Lemmatization |
|--------|----------|---------------|
| Method | Rule-based cutting | Dictionary + grammar based |
| Speed | Fast | Slower |
| Accuracy | Less accurate | More accurate |
| Output | May not be a real word | Usually a valid word |

---

### Conclusion:
- **Stemming** is useful when speed is more important.
- **Lemmatization** is better when accuracy and correct word meaning are important.

In production-level NLP systems, lemmatization is often preferred when language quality matters.

```python
import re
import string
from collections import Counter
import pandas as pd
```

```python
KEEP_SHORT_WORDS = {"no", "not"}

def remove_urls_and_emails(text):
    """Remove URLs and email-like patterns."""
    text = re.sub(r'https?://\S+|www\.\S+', ' ', text)   # URLs
    text = re.sub(r'\S+@\S+', ' ', text)                 # Emails
    return text

def remove_numbers(text):
    """Remove numbers from text."""
    return re.sub(r'\d+', ' ', text)

def normalize_repeated_characters(text):
    """
    Reduce repeated characters.
    Example:
    soooo -> soo
    goooood -> good
    nooooo -> noo
    """
    return re.sub(r'(.)\1{2,}', r'\1\1', text)

def remove_punctuation(text):
    """Remove punctuation symbols."""
    return text.translate(str.maketrans('', '', string.punctuation))

def remove_extra_spaces(text):
    """Remove extra spaces."""
    return re.sub(r'\s+', ' ', text).strip()

def tokenize_and_filter(text):
    """
    Tokenize text and remove very short tokens (length <= 2),
    except meaningful words like 'no' and 'not'.
    """
    tokens = text.split()
    filtered_tokens = []
    
    for token in tokens:
        if token in KEEP_SHORT_WORDS:
            filtered_tokens.append(token)
        elif len(token) > 2:
            filtered_tokens.append(token)
    
    return filtered_tokens
```

```python
def preprocess_text(text):
    """
    Preprocess raw text into clean tokens and clean sentence.
    
    Steps:
    1. Handle invalid / empty input
    2. Lowercase conversion
    3. Remove URLs and emails
    4. Remove numbers
    5. Normalize repeated characters
    6. Remove punctuation
    7. Remove extra spaces
    8. Remove short tokens (<=2), except 'no' and 'not'
    """
    
    
    if not isinstance(text, str):
        return {
            "original_text": text,
            "cleaned_tokens": [],
            "cleaned_sentence": ""
        }
    
    if text.strip() == "":
        return {
            "original_text": text,
            "cleaned_tokens": [],
            "cleaned_sentence": ""
        }
    
    text = text.lower()

    text = remove_urls_and_emails(text)

    text = remove_numbers(text)

    text = normalize_repeated_characters(text)

    text = remove_punctuation(text)

    text = remove_extra_spaces(text)

    tokens = tokenize_and_filter(text)

    cleaned_sentence = " ".join(tokens)
    
    return {
        "original_text": text,
        "cleaned_tokens": tokens,
        "cleaned_sentence": cleaned_sentence
    }
```

```python
def token_analytics(tokens):
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    
    if total_tokens == 0:
        avg_token_length = 0
    else:
        avg_token_length = sum(len(token) for token in tokens) / total_tokens
    
    return {
        "total_tokens": total_tokens,
        "unique_tokens": unique_tokens,
        "avg_token_length": round(avg_token_length, 2)
    }
```

```python
analytics_data = []

for original, result in zip(test_sentences, results):
    tokens = result["cleaned_tokens"]
    stats = token_analytics(tokens)
    
    analytics_data.append({
        "Original Text": original,
        "Cleaned Sentence": result["cleaned_sentence"],
        "Total Tokens": stats["total_tokens"],
        "Unique Tokens": stats["unique_tokens"],
        "Average Token Length": stats["avg_token_length"]
    })

df_analytics = pd.DataFrame(analytics_data)
df_analytics
```

```python
def full_pipeline(text_list):
    """
    Full NLP preprocessing pipeline for a list of sentences.
    
    Returns:
    {
        "tokens": [...],
        "clean_sentences": [...]
    }
    """
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        result = preprocess_text(text)
        all_tokens.extend(result["cleaned_tokens"])
        clean_sentences.append(result["cleaned_sentence"])
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```
