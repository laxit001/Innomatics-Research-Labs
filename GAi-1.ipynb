## Task 4: ChatPromptTemplate System

Build a chat-based prompt system using roles:
- teacher
- interviewer
- motivator

```python
def get_chat_prompt(role, topic):
    role_behaviors = {
        "teacher": "You are a helpful teacher who explains concepts simply and clearly.",
        "interviewer": "You are a professional interviewer who asks thoughtful questions.",
        "motivator": "You are an inspiring motivator who encourages and energizes the learner."
    }

    system_message = role_behaviors.get(role, "You are a helpful AI assistant.")

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Topic: {topic}")
    ])

    return chat_template.format_messages(topic=topic)

role = "teacher"
topic = "Neural Networks"

chat_output = get_chat_prompt(role, topic)

print("Task 4 Output:\n")
for message in chat_output:
    print(f"{message.type.upper()}: {message.content}")
```

## Task 5: Input Validation Layer

Validate:
- audience ∈ [beginner, intermediate, expert]
- tone ∈ [formal, casual, fun]

```python
def validate_inputs(audience, tone):
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    if audience not in valid_audiences:
        raise ValueError(f"Invalid audience: {audience}. Allowed values: {valid_audiences}")

    if tone not in valid_tones:
        raise ValueError(f"Invalid tone: {tone}. Allowed values: {valid_tones}")

    return True

print("Task 5 Output:")
try:
    validate_inputs("beginner", "fun")
    print("Valid inputs accepted.")
except ValueError as error:
    print(error)

try:
    validate_inputs("kids", "friendly")
except ValueError as error:
    print("Validation Error:", error)
```

## Task 6: Prompt Generator App

Create a function:
`generate_prompt(topic, audience, tone, style)`

Supported styles:
- teaching
- interview
- storytelling

```python
def generate_prompt(topic, audience, tone, style):
    validate_inputs(audience, tone)

    style_templates = {
        "teaching": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} teaching style"
        ),
        "interview": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Ask interview questions about {topic} for {audience} in a {tone} style"
        ),
        "storytelling": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} storytelling style"
        )
    }

    if style not in style_templates:
        raise ValueError("Invalid style. Choose from: teaching, interview, storytelling")

    return style_templates[style].format(
        topic=topic,
        audience=audience,
        tone=tone
    )

print("Task 6 Output:\n")
generated = generate_prompt(
    topic="Neural Networks",
    audience="beginner",
    tone="fun",
    style="storytelling"
)
print(generated)
```

## Task 7: Template Reusability Test

Use one template with 5 different inputs to prove reusability.

```python
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

inputs = [
    {"topic": "AI", "audience": "beginner", "tone": "formal"},
    {"topic": "Python", "audience": "intermediate", "tone": "casual"},
    {"topic": "Data Science", "audience": "expert", "tone": "formal"},
    {"topic": "Robotics", "audience": "beginner", "tone": "fun"},
    {"topic": "Cloud Computing", "audience": "intermediate", "tone": "casual"}
]

print("Task 7 Outputs:\n")
for item in inputs:
    result = reusable_template.format(
        topic=item["topic"],
        audience=item["audience"],
        tone=item["tone"]
    )
    print(result)
```
